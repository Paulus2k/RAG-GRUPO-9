```python
documentos = []

for archivo in archivos_txt:
    with open(archivo, "r", encoding="utf-8") as f:
        texto = f.read()
        documentos.append({
            "archivo": archivo,
            "texto": texto
        })

print("Cantidad de documentos leídos:", len(documentos))
print(documentos[0]["archivo"])
print(documentos[0]["texto"][:300])
```

**Extraer el tema y el contenido de cada documento**

```python
registros = []

for doc in documentos:
    texto = doc["texto"]
    lineas = texto.splitlines()

    tema = ""
    contenido = ""
    leyendo_contenido = False
    contenido_lineas = []

    for linea in lineas:
        linea = linea.strip()

        if linea.startswith("Tema:"):
            tema = linea.replace("Tema:", "").strip()

        elif linea.startswith("Contenido:"):
            leyendo_contenido = True
            contenido_inicial = linea.replace("Contenido:", "").strip()
            if contenido_inicial:
                contenido_lineas.append(contenido_inicial)

        elif leyendo_contenido:
            if linea:
                contenido_lineas.append(linea)

    contenido = " ".join(contenido_lineas)

    registros.append({
        "archivo": doc["archivo"],
        "tema": tema,
        "contenido": contenido
    })

print("Cantidad de registros:", len(registros))
print("Primer tema:", registros[0]["tema"])
print("Primer contenido:", registros[0]["contenido"][:300])
```

Generar el archivo SQL nuevo

```python
def escapar_sql(texto):
    texto = texto.replace("\\", "\\\\")
    texto = texto.replace("'", "\\'")
    return texto

sql = []
sql.append("DROP TABLE IF EXISTS documentos;")
sql.append("""
CREATE TABLE documentos (
    id INT AUTO_INCREMENT PRIMARY KEY,
    tema VARCHAR(100),
    contenido TEXT
);
""")

sql.append("INSERT INTO documentos (tema, contenido) VALUES")

valores = []
for r in registros:
    tema = escapar_sql(r["tema"])
    contenido = escapar_sql(r["contenido"])
    valores.append(f"('{tema}', '{contenido}')")

sql.append(",\n".join(valores) + ";")

contenido_sql_nuevo = "\n".join(sql)

with open("documentos_actualizados.sql", "w", encoding="utf-8") as f:
    f.write(contenido_sql_nuevo)

print("Archivo SQL generado correctamente.")
print("Nombre del archivo: documentos_actualizados.sql")
```

# *Función para chunking*
